```python
# 2. Read or convert the columns â€˜Created Dateâ€™ and Closed Dateâ€™ to datetime datatype
df['Created Date'] = pd.to_datetime(df['Created Date'])
```

```python
df['Closed Date'] = pd.to_datetime(df['Closed Date']) #YYYY-MM-DD 
```

```python
df.head()
```

```python
df['Request_Closing_Time'] = df['Closed Date']-df['Created Date']
```

```python
Request_Closing_Time=[]
for x in (df["Closed Date"]-df["Created Date"]):
    close=x.total_seconds()/60
    Request_Closing_Time.append(close)
    
df["Request_Closing_Time"]=Request_Closing_Time
```

```python
import numpy as np
```

```python
# Applying log transformation
dataset={}
for i in df['Complaint Type'].unique():
    dataset[i]= np.log(df[df['Complaint Type']==i]['Request_Closing_Time'])
```

```python
dataset['Blocked Driveway'].hist()
```

```python
from scipy.stats import f_oneway
stat,p = f_oneway(dataset['Noise - Street/Sidewalk'], dataset['Blocked Driveway'], dataset['Illegal Parking'], dataset['Derelict Vehicle'], 
                dataset['Noise - Commercial'])

alpha=0.05
if p>0.05:
    print('Null Hypothesis is accepted')
else:
    print('Null hypothesis is rejected')
```
